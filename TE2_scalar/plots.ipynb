```python
import numpy as np
import matplotlib.pyplot as plt

# Variable a modifier avant execution
nomfichier = "data/euler_0.1.txt" # <== fichier de donnees
#nomfichier2 = "data/euler_0.2.txt"
#nomfichier3 = "sin_centered.txt"
#nomfichier = "logistique.txt"                     # <== fichier de donnees

# Récupération du nom de la fonction tabulée
fonction_tab = nomfichier[ : nomfichier.rfind(".")]
#fonction_tab2 = nomfichier2[ : nomfichier2.rfind(".")]
#fonction_tab3 = nomfichier3[ : nomfichier3.rfind(".")]
nomgraph = fonction_tab+".pdf" #  idem fic. entree avec extension .pdf

try:      # ouverture du fichier de donnees
  fich = open(nomfichier, "r")
#  fich2 = open (nomfichier2, "r")
#  fich3 = open (nomfichier3, "r")
except:
  print(" ! Impossible d'ouvrir le fichier %s."%nomfichier )
  print(" ! Verifier le chemin.\n")
  exit()

# Lecture des lignes d'entete
ligne = fich.readline().replace("\n","") # sans caractere de fin de ligne
entete= [ ]
while( ligne[0] == "#" ):
   entete.append( ligne[1:] )
   ligne =  fich.readline().replace("\n","")
fich.close()
print( " > fichier %s : %d lignes d'entete"%( nomfichier, len(entete) ) )

# concatenation des entetes en une seule ligne de titre :
titre = " ".join(entete)

# Lecture des donnees du fichier :
tab = np.loadtxt( nomfichier, dtype="float")
#tab2 = np.loadtxt( nomfichier2, dtype="float")
#tab3 = np.loadtxt( nomfichier3, dtype="float")

# Creation des listes d'abscisses/ordonnees : 1ere et 2eme colonnes
x  = tab[:,0]
y  = tab[:,1]

x2 = tab[:,0]
y2 = tab[:,2]

x3 = tab[:,0]
y3 = tab[:,3]

# Creation du graphique
figure, (fig1, fig2) = plt.subplots(2)
figure.tight_layout(pad=4.0)

fig1.set_title( " Solutions Analytique et numérique \n" + titre, fontsize=8)
fig1.plot( x,y, label="fonction Num" )
fig1.plot( x2,y2, label="fonction Th" )
fig1.legend()
fig1.set_xlabel("temps (en SI)")                                   # <== variable d'abscisse
fig1.set_ylabel("y (en SI)")                                   # <== variable d'ordonnee
fig1.grid()

fig2.set_title( " Ecart entre les deux solutions \n" + titre, fontsize=8)
fig2.plot( x3,y3, label=fonction_tab )
fig2.set_xlabel("temps (en SI)")                                   # <== variable d'abscisse
fig2.set_ylabel("y (en SI)")                                   # <== variable d'ordonnee
fig2.grid()


#plt.plot(x, y, label=fonction_tab)                # <== choix de la legende
#plt.loglog(x,y, label=fonction_tab)
#plt.loglog(x2,y2, label=fonction_tab2)
#plt.loglog(x3,y3, label=fonction_tab3)


plt.autoscale(enable=True, axis='both', tight=None)

#plt.legend()
fig = plt.gcf()
fig.savefig(nomgraph) # sauvegarde du fichier au format PDF
print(" < Le graphique %s a ete cree."%(nomgraph) )
print("   \tVous pouvez le visualiser avec :  okular %s"%(nomgraph))

plt.show()
plt.clf() # nettoyage ('clear figure')

#print(entete[1])
euler_1 = float(entete[1][ entete[1].rfind(":") + 1 :])
#print(max1)
```

## Observations Generales

* On observe que lorsqu'on diminue le pas d'integration on diminue l'erreur maximale absolue entre la solution numérique et la solution analytique.

* On voit que l'integrtion à partir de methodes d'integration d'ordre superieur permet de diminuer l'erreur maximale considérablement. Par exemple la methode d'Euler a une erreur qui semble linéaire avce le pas. Alors que la methode des points milieux nous permet d'avoir une erreur qui varie comme le carré du pas et avec celle de RK4 on a une erreur en puissance 4 du pas. 

* Une étude plus détaillé des erreurs en fonction des pas d'integration nous permettrait d'optimiser cette grandeur dans l'objectif de minimiser l'erreur globale.

## Optimisation du pas d'integration
